```python
#import libraries
import pandas as pd
import numpy as np
pd.set_option('display.expand_frame_repr', False)
```

```python
# import file
df = pd.read_pickle("/Users/ilseeb/Desktop/Postgraduate/Train Punctuality Data/final_df.pkl")
```

```python
df.head()
```

```python
# drop columns that will not be used for analysis

df = df.drop(columns=["TRAIN_NO", "RELATION", "PTCAR_NO", "LINE_NO_DEP"
                    , "RELATION_DIRECTION", "LINE_NO_ARR", "PLANNED_DATE_ARR"
                    , "PLANNED_DATE_DEP", "REAL_DATE_ARR", "REAL_DATE_DEP"
                    ,"PTCAR_LG_NM_NL"])
```

```python
df.head()
```

```python
#check which columns have nulls

df.describe(include="all")
```

```python
# create a list of our conditions
conditions = [
    (df["hour_time_dep"] < 6),
    (df["hour_time_dep"] >= 6) &  (df["hour_time_dep"] < 12),
    (df["hour_time_dep"] >= 12) &  (df["hour_time_dep"] < 18),
    (df["hour_time_dep"] >= 18),
    ]

# create a list of the values we want to assign for each condition
values = ["night", "morning", "afternoon", "night"]

# create a new column and use np.select to assign values to it using our lists as arguments
df["time_of_day"] = np.select(conditions, values)

# display updated DataFrame
df.head()
```

```python
# set option for pandas to display all columns at once
pd.set_option('display.max_columns', 500)
```

```python
# analyze rows where the temperature is null
df[df["temperature"].isnull()]
```

```python
# delete all rows where the temperature is null
df.dropna(subset=["temperature"], inplace=True)
```

```python
df.dropna(subset=["PLANNED_TIME_ARR"], inplace=True)
```

```python
df = df.drop(columns=["REAL_TIME_ARR", "REAL_TIME_DEP", "PLANNED_TIME_ARR"
                    , "PLANNED_TIME_DEP", "DELAY_ARR", "departure", "destination"
                    , "hour_time_dep"])
```

```python
# change data types
df["TRAIN_SERV"] = df["TRAIN_SERV"].astype(str)
df["DELAY_DEP"] = pd.to_numeric(df["DELAY_DEP"])
df["delayed"] = df["delayed"].astype(bool)
df["destination_region"] = df["destination_region"].astype(str)
df["departure_region"] = df["departure_region"].astype(str)
df["time_of_day"] = df["time_of_day"].astype(str)
```

```python
df.dtypes
```

```python
df.rename(columns={"TRAIN_SERV":"train_serv", "DELAY_DEP":"delay_dep"}, inplace=True)
```

```python
df["date"] = df["day"].astype(str) + "-" + df["month"].astype(str) + "-" + df["year"].astype(str)
```

```python
df["date"] = df["date"].apply(pd.to_datetime)
```
