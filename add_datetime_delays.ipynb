```python
#import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/ilseeb/Downloads/Train Punctuality Data/final_dataset.csv")
```

```python
df.head()
```

```python
df["year"] = df["DATDEP"].apply(lambda x: x[5:])
```

```python
df["month"] = df["DATDEP"].apply(lambda x: x[2:5])
```

```python
df["day"] = df["DATDEP"].apply(lambda x: x[:2])
```

```python
df["month"].unique()
```

```python
df["month"] = df["month"].map({"JAN": "01", "FEB": "02", "MAR": "03", "APR": "04", "MAY": "05", "JUN": "06", "JUL": "07", "AUG": "08", "SEP": "09", "OCT": "10", "NOV": "11", "DEC": "12"})
```

```python
df.head()
```

```python
df["hour_time_dep"] = df["REAL_TIME_DEP"].astype("datetime64").dt.round("H").dt.hour
```

```python
df.head()
```

```python
df.dtypes
```

```python
df_regions = pd.read_excel("/Users/ilseeb/Downloads/Train Punctuality Data/train_stations_regions_2.xlsx")
```

```python
df_all = pd.merge(df, df_regions, how="left", left_on = ["destination"], right_on = ["train_stations" ])
```

```python
df_all.head()
```

```python
df_weather = pd.read_csv("/Users/ilseeb/Downloads/Train Punctuality Data/weather_data.csv")
```

```python
df_weather.dtypes
```

```python
df_all["year"] = df_all["year"].astype(int)
df_all["month"] = df_all["month"].astype(int)
df_all["day"] = df_all["day"].astype(int)
```

```python
df_all.dtypes
```

```python
df_all_2 = pd.merge(df_all, df_weather, how="left", left_on = ["year", "month", "day", "hour_time_dep"], right_on = ["year", "month", "day", "time"])
```
