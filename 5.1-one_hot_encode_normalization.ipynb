```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
#read pickle
X_train = pd.read_pickle("/Users/ilseeb/Desktop/Postgraduate/Train Punctuality Data/train_test_data/X_train_over.pkl")
```

```python
X_train.columns
```

```python
one_hot_encoded_data = pd.get_dummies(X_train, columns = ['train_serv', 'month'
        , 'destination_region', 'departure_region',
       'time_of_day', 'day_of_week'])
```

```python
one_hot_encoded_data.head()
```

```python
df_to_normalize = one_hot_encoded_data[["temperature","relative_humidity"
                                           ,"wind_speed", "wind_direction"]]
```

```python
min_max = preprocessing.MinMaxScaler()
scaled_df = min_max.fit_transform(df_to_normalize.values)
normalized_df = pd.DataFrame(scaled_df, columns=["temperature","relative_humidity"
                                           ,"wind_speed", "wind_direction"])
```

```python
one_hot_encoded_data["temperature"] = normalized_df["temperature"]
one_hot_encoded_data["relative_humidity"] = normalized_df["relative_humidity"]
one_hot_encoded_data["wind_speed"] = normalized_df["wind_speed"]
one_hot_encoded_data["wind_direction"] = normalized_df["wind_direction"]
```

```python
one_hot_encoded_data.head()
```
