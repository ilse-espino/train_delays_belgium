```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#download first file
df = pd.read_csv("/Users/ilseeb/Downloads/Train Punctuality Data/2014/Data_raw_punctuality_201401.csv")
```

```python
#count how much data we have
df.count()
```

```python
df.head(100)
```

```python
#removed the trains which ended in Brussels
df['PTCAR_LG_NM_NL'][pd.isna(df.REAL_TIME_DEP)]
```

```python
#get final destination by grouping per train and date of departure
df['destination'] = df.groupby(['DATDEP', 'TRAIN_NO'], sort=False)['PTCAR_LG_NM_NL'].transform('last')

```

```python
df.head(1220)
```

```python
#get only rows where the train departed from BRUSSEL-CENTRAAL
df2 = df[df.PTCAR_LG_NM_NL == 'BRUSSEL-CENTRAAL']
```

```python
df2.count()
```

```python
#add column to define if train was delayed or not
df2['delayed'] = np.where(df2['DELAY_DEP'] > 360, True, False)
```
